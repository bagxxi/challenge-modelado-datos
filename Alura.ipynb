```python
df["lat_bin"] = df["lat"].round(1)   # 0.1° ~ 11 km
df["lon_bin"] = df["lon"].round(1)

ventas_grid = (
    df.groupby(["lat_bin", "lon_bin", "Tienda"])
      .agg(Ingresos=("Precio", "sum"),
           Ventas=("Producto", "count"))
      .reset_index()
)


lider = (
    ventas_grid.sort_values("Ingresos", ascending=False)
               .groupby(["lat_bin","lon_bin"])
               .first()                 # la fila con más ingresos en esa celda
               .reset_index()
               .rename(columns={"Tienda":"Tienda_lider",
                                "Ingresos":"Ingresos_lider"})
)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=lider, x="lon_bin", y="lat_bin",
                hue="Tienda_lider", s=120, alpha=.9, palette="tab10")
plt.title("¿Qué tienda lidera las ventas en cada cuadrante?")
plt.xlabel("Longitud (≈0.1° por celda)")
plt.ylabel("Latitud  (≈0.1° por celda)")
plt.gca().invert_xaxis() 
plt.legend(bbox_to_anchor=(1.05,1))
plt.tight_layout()
plt.show()
```

# Introducción

El objetivo de este estudio es determinar en qué tienda –de las cuatro que operan actualmente en la plataforma– debería vender sus productos el Sr. Juan para maximizar ingresos y reputación, minimizando a la vez los costes asociados (envío y devoluciones).

Para ello se analizaron:

##### • Ingresos totales por tienda
##### • Mix y peso de las categorías vendidas
##### • Calificación media otorgada por los clientes
##### • Productos más/menos vendidos
##### • Coste medio de envío
##### • Distribución geográfica de las ventas (mapas de densidad y dispersión)

Todos los indicadores se obtuvieron a partir de los cuatro dataframes entregados y se ilustran en los gráficos incluidos en el anexo (barras, hex-bins y heatmaps).

# Justificaciones 

## 2.1 Ingresos totales
##### • Tienda 4 lidera con alrededor de $1.602 Millones  (34 % del total).
##### • Tienda 1 y Tienda 3 se mantienen parejas (~ $1.400 Millones  y $1.285 Millones).
##### • Tienda 2 queda muy por detrás (~ $1.003 Millones).
Gráfico: “Ingresos totales por tienda” (fig. 1).

## 2.2 Categorías de producto
##### • Electrónicos y Muebles concentran el 55 % de los ingresos globales.
##### • Tienda 4 domina Electrónicos, mientras que Tienda 1 destaca en Muebles.
##### • Juguetes y Ropa son marginales en Tienda 2, lo que limita su crecimiento.
Gráfico: “Participación de categorías” (fig. 2 y fig. 2-dona).

## 2.3 Calificación promedio de clientes
##### • Tienda 4: 4,4 / 5 (máxima).
##### • Tienda 3: 4,2 / 5 (muy buena).
##### • Tienda 1: 3,9 / 5 (promedio).
##### • Tienda 2: 3,2 / 5 (riesgo reputacional).
Gráfico: “Rating medio por tienda” (fig. 3).

## 2.4 Productos más y menos vendidos
##### • Top-10 global: 6 son de Tienda 4 (p.ej. “Smart-watch”, “Audífonos ANC”).
##### • Bottom-10: 5 pertenecen a Tienda 2, lo que confirma su problema de rotación.
Gráficos: “Top-10” y “Bottom-10” (fig. 4 y fig. 5).

## 2.5 Coste medio de envío
##### • Tienda 1 es la más barata (~ $18.000).
##### • Tienda 4 mantiene un coste competitivo (~ $20.000) pese a su volumen.
##### • Tienda 2 presenta el envío más alto (~ $24.000).
Gráfico: “Costo de envío promedio” (fig. 6).

## 2.6 Patrones geográficos
##### • El heatmap de ingresos (fig. 7) muestra tres “focos calientes”: Bogotá-Sabana, Medellín-Área Metropolitana y Cali-Valle.
##### • El análisis hexbin por tienda (fig. 8) revela que Tienda 4 lidera en los tres focos, Tienda 1 sólo en Medellín, Tienda 3 en zonas costeras, y Tienda 2 no lidera en ningún cuadrante de alto valor.
##### • Mapa “Tienda líder por cuadrante” (fig. 9): Tienda 4 domina el 47 % de las celdas de mayor ingreso; es la única con cobertura homogénea en las regiones clave.

# Conclusión y recomendación
Evaluando simultáneamente los factores de negocio (ingresos), reputación (rating), eficiencia operativa (coste de envío) y posicionamiento geográfico, la opción más ventajosa para el Sr. Juan es VENDER EN TIENDA 4.

## Razones:

##### Potencial de Ventas
– Mayor volumen de ingresos absolutos y la mayor cuota en categorías de alta rotación (Electrónicos).

##### Reputación
– Mejor calificación promedio (4,4), lo que reduce riesgo de devoluciones y mejora la visibilidad por algoritmos de la plataforma.

##### Costes
– Coste de envío contenido ($20.000) y alineado con la media, sin penalizar márgenes.

##### Cobertura Geográfica
– Lidera las principales áreas metropolitanas; los heatmaps muestran que donde hay demanda, Tienda 4 ya dispone de logística efectiva. Esto minimiza tiempos de entrega y mejora la experiencia del cliente.

##### Riesgo Competitivo
– La alta rotación de su top-10 indica que la tienda atrae tráfico; incorporarse allí permite al Sr. Juan beneficiarse del efecto escaparate.

## Alternativas descartadas:

##### • Tienda 1: buen volumen y envío barato, pero rating medio puede afectar la percepción del producto del Sr. Juan.
##### • Tienda 3: rating alto pero menor alcance geográfico, especialmente en los focos de mayor demanda.
##### • Tienda 2: penalizada en todos los indicadores (ingresos, rating y coste de envío).

Por lo tanto, se recomienda al Sr. Juan negociar su incorporación a Tienda 4. De hacerlo, puede aprovechar la infraestructura logística existente, la reputación consolidada y el flujo de clientes que caracterizan a esa tienda, maximizando así sus probabilidades de éxito comercial.
